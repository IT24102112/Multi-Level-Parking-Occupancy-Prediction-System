```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cleaning Processes
# Rename columns for easier access based on the head() output seen earlier
# The current columns seem to be raw data without headers in the CSV
df.columns = ['vehicle_id', 'vehicle_type', 'type', 'arrival_time', 'departure_time']

# Convert to datetime
df['arrival_time'] = pd.to_datetime(df['arrival_time'])
df['departure_time'] = pd.to_datetime(df['departure_time'])

# Filter 1: Only 'COUNTER' type (ignore SPEC. RESERVATION)
df = df[df['type'] == 'COUNTER'].copy()

# Filter 2: Drop 'Mortor Cycle' and 'three wheelar'
# Using case-insensitive matching to be safe, but specifically targeting the requested strings
exclude_types = ['Mortor Cycle', 'three wheelar']
df = df[~df['vehicle_type'].isin(exclude_types)].copy()

# 2. Capacity Estimation
# To calculate availability, we need the total capacity.
# We'll calculate the number of concurrent vehicles at any given time.
events = []
for _, row in df.iterrows():
    events.append((row['arrival_time'], 1))
    events.append((row['departure_time'], -1))

events.sort()

current_occupancy = 0
max_occupancy = 0
occupancy_timeline = []

for time, change in events:
    current_occupancy += change
    occupancy_timeline.append((time, current_occupancy))
    if current_occupancy > max_occupancy:
        max_occupancy = current_occupancy

# We'll define Capacity as the historical peak occupancy
capacity = max_occupancy
print(f'Estimated Total Capacity (based on peak): {capacity}')

# 3. Create Time Series for LSTM (15-min intervals)
df_resampled = pd.DataFrame(occupancy_timeline, columns=['timestamp', 'occupancy'])
df_resampled.set_index('timestamp', inplace=True)
df_resampled = df_resampled.resample('15T').last().ffill().fillna(0)

# Calculate Percentage Availability
df_resampled['availability_pct'] = ((capacity - df_resampled['occupancy']) / capacity) * 100

display(df_resampled.head())
```

```python
# Average peak times in all years, months, and weeks
df_resampled['year'] = df_resampled.index.year
df_resampled['month'] = df_resampled.index.month
df_resampled['week'] = df_resampled.index.isocalendar().week
df_resampled['hour'] = df_resampled.index.hour

plt.figure(figsize=(15, 10))

# Yearly Peak Trends
plt.subplot(3, 1, 1)
sns.lineplot(data=df_resampled, x='year', y='occupancy', estimator='max')
plt.title('Max Peak Occupancy per Year')

# Monthly Seasonality
plt.subplot(3, 1, 2)
sns.barplot(data=df_resampled, x='month', y='occupancy', estimator='mean')
plt.title('Average Occupancy per Month')

# Weekly/Daily Peak (Average by Hour of Day)
plt.subplot(3, 1, 3)
sns.lineplot(data=df_resampled, x='hour', y='occupancy', estimator='mean')
plt.title('Average Hourly Occupancy Profile (Peak Times)')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Prepare data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_resampled[['availability_pct']])

def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset)-look_back-1):
        X.append(dataset[i:(i+look_back), 0])
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)

# Look back 4 hours (16 intervals of 15 mins)
look_back = 16
X, y = create_dataset(scaled_data, look_back)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Build LSTM
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(look_back, 1)),
    Dropout(0.2),
    LSTM(50),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
# Note: Reduced epochs for demonstration; increase for better results
model.fit(X, y, epochs=5, batch_size=32, verbose=1)

print("Model trained. It is now ready to predict availability for the next 15-minute window.")
```

```python
# SECTION 1: DATA CLEANING
import pandas as pd
import numpy as np

# Defining column names
df.columns = ['vehicle_id', 'vehicle_type', 'type', 'arrival_time', 'departure_time']

# Convert timestamps to datetime objects
df['arrival_time'] = pd.to_datetime(df['arrival_time'])
df['departure_time'] = pd.to_datetime(df['departure_time'])

# Filter for COUNTER type only
df_clean = df[df['type'] == 'COUNTER'].copy()

# Filter out 2/3 wheelers as requested
exclude_list = ['Mortor Cycle', 'three wheelar']
df_clean = df_clean[~df_clean['vehicle_type'].isin(exclude_list)].copy()

print(f'Rows after cleaning: {len(df_clean)}')
```

```python
# SECTION 2: CAPACITY ESTIMATION
# We calculate concurrent occupancy to find the peak
events = []
for _, row in df_clean.iterrows():
    events.append((row['arrival_time'], 1))   # Arrival adds a vehicle
    events.append((row['departure_time'], -1)) # Departure removes a vehicle

events.sort()

curr, peak = 0, 0
for _, change in events:
    curr += change
    if curr > peak: peak = curr

capacity = peak
print(f'Calculated Total Capacity: {capacity} slots')
```

```python
# SECTION 4: PEAK TREND VISUALIZATION
import matplotlib.pyplot as plt
import seaborn as sns

df_resampled['year'] = df_resampled.index.year
df_resampled['month'] = df_resampled.index.month
df_resampled['hour'] = df_resampled.index.hour

fig, axes = plt.subplots(3, 1, figsize=(12, 15))

sns.lineplot(ax=axes[0], data=df_resampled, x='year', y='occupancy', estimator='max')
axes[0].set_title('Yearly Max Peak Occupancy')

sns.barplot(ax=axes[1], data=df_resampled, x='month', y='occupancy', estimator='mean')
axes[1].set_title('Average Occupancy by Month')

sns.lineplot(ax=axes[2], data=df_resampled, x='hour', y='occupancy', estimator='mean')
axes[2].set_title('Average Hourly Occupancy (Daily Cycle)')

plt.tight_layout()
plt.show()
```

```python
def predict_availability(latest_data_15min_intervals):
    """
    Expects an array of the last 16 (4 hours) 15-minute availability percentages.
    """
    if len(latest_data_15min_intervals) < 16:
        return "Error: Need at least 16 data points (4 hours) to predict."

    # Scale the input
    input_data = np.array(latest_data_15min_intervals).reshape(-1, 1)
    scaled_input = scaler.transform(input_data)

    # Reshape for LSTM [samples, time steps, features]
    X_input = scaled_input.reshape((1, 16, 1))

    # Predict
    prediction_scaled = model.predict(X_input, verbose=0)

    # Inverse transform to get percentage
    prediction_pct = scaler.inverse_transform(prediction_scaled)

    return max(0, min(100, prediction_pct[0][0]))

# Example Usage: Predict next 15 mins using the last 16 rows of our data
last_window = df_resampled['availability_pct'].tail(16).values
forecast = predict_availability(last_window)
print(f'Predicted Availability for the next 15 minutes: {forecast:.2f}%')
```

```python
# SECTION 5: LSTM MODELING
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_resampled[['availability_pct']])

def create_sequences(data, window=16):
    x, y = [], []
    for i in range(len(data)-window-1):
        x.append(data[i:(i+window), 0])
        y.append(data[i+window, 0])
    return np.array(x), np.array(y)

look_back = 16 # 4 hours of history
X_data, y_data = create_sequences(scaled_data, look_back)
X_data = X_data.reshape((X_data.shape[0], X_data.shape[1], 1))

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(look_back, 1)),
    Dropout(0.2),
    LSTM(32),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X_data, y_data, epochs=3, batch_size=64, verbose=1)

print('Model training complete for 15-minute availability prediction.')
```
